```python
import pandas as pd
df=pd.read_csv("students.csv")
print(df.head)
```

```python
# Display the dataset
print("Student Performance Data")
print(df)
```

```python
# Display dataset information
print("\nInformation")
print(df.info())

# Display statistical summary
print("\nStatistics")
print(df.describe())

# Add Total Marks
df["Total"] = df["Math"] + df["Science"] + df["English"]

# Add Average Marks
df["Average"] = df["Total"] / 3

# Display updated dataframe
print("\nData with Total and Average")
print(df)

# Find Top Scorer
top_student = df[df["Average"] == df["Average"].max()]

print("\nTop Scorer")
print(top_student)

# Students scoring more than 75 average
high_performers = df[df["Average"] > 75]

print("\nStudents with Average > 75")
print(high_performers)

# Save updated data into a new CSV file
df.to_csv("student_result.csv", index=False)

print("\nUpdated data saved as student_result.csv")
```

```python
# Display the dataset
print("Student Performance")
print(df)

# Calculate Total Marks
df["Total"] = df["Math"] + df["Science"] + df["English"]

# Calculate Average Marks
df["Average"] = df["Total"] / 3

# Display updated data
print("\nStudent Result")
print(df)

# Find Top Scorer
print("\nTop Scorer")
top = df[df["Average"] == df["Average"].max()]
print(top)

# Students who scored above 75 average
print("\nStudents with Average > 75")
print(df[df["Average"] > 75])

# Find highest marks in Math
print("\nHighest Math Mark:", df["Math"].max())

# Find lowest marks in Science
print("Lowest Science Mark:", df["Science"].min())

# Average marks of English
print("Average English Mark:", df["English"].mean())

# Save the updated dataframe
df.to_csv("student_result.csv", index=False)

print("\nData saved successfully.")
```

```python
# Display first 3 records
print("\nFirst 3 Students")
print(df.head(3))

# Display last 2 records
print("\nLast 2 Students")
print(df.tail(2))

# Display column names
print("\nColumn Names")
print(df.columns)

# Display dataset information
print("\nDataset Information")
print(df.info())

# Display statistical summary
print("\nStatistical Summary")
print(df.describe())

# Students with GPA greater than 8.5
print("\nStudents with GPA > 8.5")
print(df[df["GPA"] > 8.5])

# Students from CSE Department
print("\nCSE Students")
print(df[df["Department"] == "CSE"])

# Highest GPA
print("\nHighest GPA")
print(df[df["GPA"] == df["GPA"].max()])

# Average GPA
print("\nAverage GPA")
print(df["GPA"].mean())

# Number of students in each department
print("\nStudents in Each Department")
print(df["Department"].value_counts())

# Sort students by GPA (Highest to Lowest)
print("\nStudents Sorted by GPA")
print(df.sort_values(by="GPA", ascending=False))

# Save updated data
df.to_csv("student_details.csv", index=False)

print("\nData saved successfully.")
```
